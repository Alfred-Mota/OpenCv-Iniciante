```python
import cv2
from pathlib import Path
import os
import numpy as np
import matplotlib.pyplot as plt
```

Lendo, Escrevendo e Vendo Imagem com CV2

```python
img_path = r"E:\HandsOnPython\Datasets\YOLO_CLS_WEATHER\weather\train\rain\rain8.jpg"
img = cv2.imread(img_path)
print(type(img))
print(img.shape)

#Salvando as imagens no diretorio ./data
# cv2.imwrite(str(Path().resolve().parent/'data'/'rain_output.jpg'), img)
# # cv2.imwrite(str(Path().resolve().parent/'data'/'rain_input.jpg'), img)

text = 'Rain'
cv2.putText(img, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
            1, (0, 255, 0), 2, cv2.LINE_AA)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Carregando e Vendo Video com CV2

```python
video_path = Path().resolve()/'data'/'video.mp4'

video = cv2.VideoCapture(str(video_path))

ret = True
while ret:
    ret, frame = video.read()

    if ret:
        cv2.imshow('Video', frame)
        cv2.waitKey(10)
        
video.release()
cv2.destroyAllWindows()
```

Imagens via WebCam com CV2

```python
webcam = cv2.VideoCapture(0)

while True:

    ret, frame = webcam.read()
    
    cv2.imshow('WebCam', frame)

    if((cv2.waitKey(40) & 0xFF) == ord('q')):
        
    # ord() retorna o unicode do caracter informado
    # cv2.waitKey(40) retorna um binario de 32 bits da tecla precionada
    # 0xFF é uma mascara 1111 1111
    # o operador & é equivalente a um AND

        break

webcam.release()
cv2.destroyAllWindows()
```
